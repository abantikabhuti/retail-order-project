```python
df = pd.read_csv(r"C:\Users\Lenovo\Desktop\retail order project\orders.csv", na_values=['Not Available','unknown'])
df['Ship Mode'].unique()
```

```python
#rename columns (lowercase and underscore)

df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.head()
```

```python
#derive new columns - discount, sale price, profit

df['discount'] = df['list_price']*df['discount_percent']*0.01
df['sale_price'] = df['list_price']-df['discount']
df['profit'] = df['sale_price']-df['cost_price']
df.head()
```

```python
#convert order_date to datetime

df['order_date']=pd.to_datetime(df['order_date'], format='%d-%m-%Y')
df.dtypes
```

```python
#drop cost price, sale price, discount % columns

df.drop(columns=['cost_price','list_price','discount_percent'], inplace=True)
df.head()
```

```python
#load data into sql server using replace

import sqlalchemy as sal
engine=sal.create_engine('mssql://RAKA\SQLEXPRESS/master?driver=ODBC+DRIVER+17+FOR+SQL+SERVER')
conn=engine.connect()
```

```python
df.to_sql('df_orders', con=conn, index=False, if_exists='replace')
```
